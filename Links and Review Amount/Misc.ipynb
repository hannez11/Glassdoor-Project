```python
df = pd.read_excel(r"W:\019_Glassdoor\1 Data\1 Glassdoor Links\Cineworld.xlsx")
df
```

```python
df = df[["CIQ Transaction ID", "Acquirer", "ciq", "isin", "correct"]]
df
```

```python
# https://stackoverflow.com/questions/70074621/python-pandas-pivot-table-for-multiple-columns-at-once-which-has-duplicate-val
df_out = df.set_index(['CIQ Transaction ID',
                       df.groupby(['CIQ Transaction ID'])\
           .cumcount() +1]).unstack()
df_out.columns = [f'{i}_{j}' for i, j in df_out.columns]
df_out = df_out.reset_index()
df_out
```

```python
df_issues = df_out[(df_out["correct_1"] == 0) | (df_out["correct_2"] == 0) & (df_out["ciq_1"] != df_out["ciq_2"])] #df_out["ciq_1"] != df_out["ciq_2"] -> es gibt vor allem fälle, in denen ISIN von acq und target identisch sind (weil nicht updated)
df_issues
```
